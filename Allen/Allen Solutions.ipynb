```python
import math

def calculate_area(diameter: float) -> float:
    return (diameter / 2) ** 2 * math.pi

def calculate_triangle_area(base: float) -> float:
    height = ((base**2) - ((base/2) ** 2)) ** .5
    return ((base /2 ) * height)

def square_area(square_Length: float) -> float:
    return (square_Length ** 2)

def dough_per_circle(dough: float, diameters: list[float]) -> float:
    total_area1 = sum([calculate_area(diameter) for diameter in diameters])
    return total_area1 / dough

def dough_per_triangle(dough: float, bases: list[float]) -> float:
    total_area2 = sum([calculate_triangle_area(base) for base in bases])
    return total_area2 / dough

def dough_per_square(dough: float, square_lengths: list[float]) -> float:
    total_area3 = sum([square_area(square_length) for square_length in square_lengths])
    return total_area3 / dough

small_cost_per_area = dough_per_circle(20, [15, 15])
print(f"dough per sq in of smalls is {small_cost_per_area:.2f}")

big_cost_per_area = dough_per_triangle(20, [20])
print(f"dough per sq in of large is {big_cost_per_area:.2f}")


square_cost_per_area = dough_per_square(18, [18])
print(f"dough per sq in of square is {square_cost_per_area:.2f}")



```
